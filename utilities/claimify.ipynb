```python
import nltk
nltk.download("punkt", quiet=True)

from pydantic import BaseModel

from langchain_ollama import ChatOllama
from langchain_core.prompts import ChatPromptTemplate

from pprint import pprint

from pathlib import Path
DATA_DIR = Path(r"C:\Users\arkma\Sqrt-1\putting the pro in programming\0. Ongoing Projects\claim-extraction\data")

TITLE = "Climate Change"
TEXT = (DATA_DIR / "sample-input.txt").read_text(encoding="utf-8")

SELECTION_SYSTEM = (DATA_DIR / "selection-system.txt").read_text()
SELECTION_USER = (DATA_DIR / "selection-user.txt").read_text()
```

# Sentence splitting

```python
def tokenise(text):
    return nltk.sent_tokenize(text)

def selection(sentences, title):
    model = ChatOllama(
        model='qwen:7b-chat',
        temperature=0
    )
    prompt = ChatPromptTemplate.from_messages([
        ('system', SELECTION_SYSTEM),
        ('user', SELECTION_USER)
    ])
    chain = prompt | model

    for i, sentence in enumerate(sentences):
        if i < 5:
            context = sentences[:i]
        else:
            context = ['[...]'] + sentences[i-5:i]
        context.append(sentence)
        if i > len(sentences) - 5:
            context.extend(sentences[i+1:])
        else:
            context.extend(sentences[i+1:i+6] + ['[...]'])

        reply = chain.invoke({    
            'headline': title,
            'excerpt': '\n'.join(context),
            'sentence': sentence
        })

        print('-' * 80)
        print(reply.content)
        print('-' * 80)
```

```python
sentences = tokenise(TEXT)
selection(sentences, TITLE)
```
